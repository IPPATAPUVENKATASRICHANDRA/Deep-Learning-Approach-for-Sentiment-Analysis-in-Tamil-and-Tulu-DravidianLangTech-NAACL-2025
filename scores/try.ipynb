```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Paths to your files (update these paths)
PREDICTIONS_FILE = "Code Conquerors_Tulu.tsv"  
TRUE_LABELS_FILE = "Tulu_SA_test_with_label (1).csv"

# Load data (adjust column names if needed)
# --------------------------------------------------
# Read TSV (predicted labels)
pred_df = pd.read_csv(PREDICTIONS_FILE, sep='\t')
y_pred = pred_df['Label'].tolist()  

# Read CSV (true labels)
true_df = pd.read_csv(TRUE_LABELS_FILE)
y_true = true_df['Label'].tolist()  

# Ensure the same number of samples
assert len(y_true) == len(y_pred), "Mismatch in number of true and predicted labels!"

# Calculate metrics
# --------------------------------------------------
accuracy = accuracy_score(y_true, y_pred)
macro_precision = precision_score(y_true, y_pred, average='macro')
macro_recall = recall_score(y_true, y_pred, average='macro')
macro_f1 = f1_score(y_true, y_pred, average='macro')

# Print results
print(f"Accuracy: {accuracy:.4f}")
print(f"Macro Precision: {macro_precision:.4f}")
print(f"Macro Recall: {macro_recall:.4f}")
print(f"Macro F1 Score: {macro_f1:.4f}")


# confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt 
conf_matrix = confusion_matrix(y_true, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()

```
